## Небольшой пример

Первая задача стала индикатором того, что вы плохо изучаете лекции (или просто теорию). Каких решений я только не видел :) но вот, наверное, самое простое:

```python
N = int(input())
array = [0] * N

print("Входной массив")

for i in range(N):
    array[i] = int(input())

print("Выходной массив")

for i in range(N-1, -1, -1):
    print(array[i], end = ' ')
```

По сути это задача на цикл **for**. Во время контеста я видел только одно такое элегантное решение у **Насти Зелениной**. Базовое понимание того, как работает *range* избавило бы от размышлений. Тонкий момент, что идти нужно от N-1, так как это максимальный индекс в массиве длины N.
